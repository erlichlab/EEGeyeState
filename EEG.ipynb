```python
raw_data = np.loadtxt(open("EEG.csv", "rb"), delimiter=",", skiprows=0)
```

```python
X_raw = raw_data[:,:-1]
X = (X_raw - np.mean(X_raw)) / np.std(X_raw)
y = raw_data[:,-1]
```

```python
abnormal = np.logical_or(X_raw>4800, X_raw<3800)
X_cleaned = np.copy(X_raw)
for i in range(X_cleaned.shape[1]):
    X_cleaned[abnormal[:,i]] = np.mean(X_cleaned,axis=0)[i]
```

```python
for i in range(X_cleaned.shape[1]):
    plt.plot(np.arange(X_cleaned.shape[0]),X_cleaned[:,i])
    plt.title("Trajectory of EEG signal")
    plt.xlabel("Time step")
    plt.ylabel("EEG values")
    plt.show()
```
